```python
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
df = pd.read_csv('sample_dataset.csv')
df = df.set_index('TID')
df # to make sure that the code imported the dataset properly
```

```python
def apriori_method(minimum_support):
    # first, we should generate the frequent itemset
    freq_itemset = apriori(df, min_support = minimum_support, use_colnames=True)
    # then, we should generate the association rules
    rules = association_rules(freq_itemset, metric='lift', min_threshold=1) # we could use eaither lift or confidence as metric
    return rules
```

```python
apriori_method(0.3)
```

```python
# we could export these rules into a csv file
apriori_method(0.3).to_csv('association_rules_0.2support.csv')
```
